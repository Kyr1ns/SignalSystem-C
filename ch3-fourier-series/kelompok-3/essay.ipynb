##### **Audacity**
Audacity adalah perangkat lunak gratis, open-source, dan multi-platform untuk merekam dan mengedit audio, yang tersedia untuk Windows, macOS, dan Linux. Perangkat lunak ini memungkinkan pengguna untuk merekam suara langsung, mengimpor dan memanipulasi berkas audio, serta menambahkan berbagai efek untuk menghasilkan segmen audio berkualitas profesional.
##### **Equalizer**
Preset Filter Curve Equalizer yang digunakan dalam percobaan ini adalah 3, yaitu:
- _Telephone_ dengan menerapkan high pass filter kepada frekuensi rendah sekitar 300-600 Hz dan menerapkan low pass filter kepada frekuensi tinggi ke tas 4-7 kHz.
- _Treble Boost_ dengan memperbesar frekuensi treble atas ke atas 10 kHz.
- _Bass Cut_ dengan meletakkan 200 Hz pada dB terendah dan 300 Hz pada 0 dB.
##### **High Pass Filter**
High-Pass Filter (HPF) adalah filter yang meloloskan frekuensi high (tinggi) frekuensi yang kita kehendaki, sehingga frekuensi low (rendah) dibawahnya disaring (filter) dan dihilangkan. Untuk menggunakan filter ini di audacity sendiri, nantinya akan menentukan cut off frekuensi yang sebagai batas filter frekuensi dan juga roll off (db per octave) untuk menentukan seberapa frekuensi menjadi lemah.
##### **Low Pass Filter**
Low-Pass Filter (LPF) adalah filter yang meloloskan frekuensi low (rendah) frekuensi yang kita kehendaki, sehingga frekuensi high (tinggi) diatasnya disaring (filter) dan dihilangkan. Untuk menggunakan filter ini di audacity sendiri, nantinya akan menentukan cut off frekuensi yang sebagai batas filter frekuensi dan juga roll off (db per octave) untuk menentukan seberapa frekuensi menjadi lemah.

### Deret Fourier (Fourier Series)

Deret Fourier **(Fourier Series)** merupakan metode matematis untuk menyatakan suatu fungsi periodik sebagai penjumlahan dari gelombang sinus dan cosinus. Prinsip dasarnya adalah bahwa sinyal periodik apa pun, seberapa kompleks pun bentuknya, dapat diuraikan menjadi kombinasi harmonik-harmonik sederhana dengan frekuensi kelipatan dari frekuensi dasar. Dengan cara ini, gelombang kotak, segitiga, maupun sawtooth sebenarnya dapat direpresentasikan sebagai gabungan banyak gelombang sinusoidal.

Secara umum, sebuah fungsi periodik \( f(t) \) dengan periode \( T \) dapat dituliskan dalam bentuk deret Fourier sebagai:

$$
f(t) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos\left(\frac{2 \pi n}{T} t\right) + b_n \sin\left(\frac{2 \pi n}{T} t\right) \right]
$$

dengan koefisien-koefisien Fourier yang didefinisikan sebagai:

$$
a_0 = \frac{1}{T} \int_{0}^{T} f(t)\, dt
$$

$$
a_n = \frac{2}{T} \int_{0}^{T} f(t) \cos\left(\frac{2 \pi n}{T} t\right)\, dt
$$

$$
b_n = \frac{2}{T} \int_{0}^{T} f(t) \sin\left(\frac{2 \pi n}{T} t\right)\, dt
$$

Dalam hal ini, \( a_0 \) merepresentasikan nilai rata-rata (atau komponen DC) dari sinyal, sedangkan \( a_n \) dan \( b_n \) menggambarkan besarnya kontribusi masing-masing komponen cosinus dan sinus. Semakin banyak suku deret Fourier yang digunakan, maka representasi grafiknya akan semakin mendekati bentuk fungsi periodik asli.

### Gelombang Sawtooth
Gelombang **sawtooth** atau gelombang gigi gergaji adalah salah satu bentuk gelombang periodik yang naik secara linear dari nilai minimum ke maksimum, lalu turun secara tiba-tiba kembali ke nilai awalnya. Sifat inilah yang membuatnya menyerupai gigi gergaji. Secara matematis, gelombang sawtooth dengan amplitudo \( A \) dan periode \( T \) dapat dinyatakan sebagai:

$$
x(t) = \frac{2A}{T} (t \bmod T) - A
$$

Persamaan tersebut menunjukkan bahwa dalam satu periode, sinyal akan naik linear dari \( -A \) hingga \( +A \) sebelum kembali ke \( -A \) secara mendadak.

Pada kode di bawah, kita akan membuat gelombang atau grafik sawtooth menggunakan python dengan matplotlib dan scipy.

```python
import matplotlib.pyplot as plt
from sawtooth import generate_sawtooth

t, x, freq =generate_sawtooth(5, 1, 1000)
plt.plot(t, x)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title(f"Sawtooth Wave - {freq} Hz")
plt.grid(True)
plt.show()
```

### Penguraian Menggunakan Representasi Fourier 

Kita akan mencoba menguraikan grafik sawtooth menggunakan fungsi fourier_series() yang telah kami buat dengan modifikasi kode dari web.

```python
from fourier import fourier_series_sawtooth

fourier_series_sawtooth(freq=5, duration=1, sample_rate=2000, terms=10);
fourier_series_sawtooth(freq=5, duration=1, sample_rate=2000, terms=100);
fourier_series_sawtooth(freq=5, duration=1, sample_rate=2000, terms=1000);
```

### Kesimpulan
Deret Fourier berhasil merepresentasikan gelombang sawtooth, dengan tingkat akurasi yang meningkat tajam seiring bertambahnya jumlah suku (dari 10 ke 1000). Semakin banyak suku digunakan, bentuk gelombang aproksimasi semakin mirip dengan garis lurus gelombang sawtooth ideal. Namun, pada titik diskontinuitas (lompatan tajam), selalu terjadi Fenomena Gibbs—sebuah overshoot dan undershoot kecil yang tidak hilang meskipun jumlah suku ditingkatkan. Peningkatan suku hanya membuat Fenomena Gibbs menjadi lebih sempit dan lebih dekat ke titik diskontinuitas, menunjukkan bahwa gelombang sawtooth membutuhkan sejumlah harmonik tak terhingga untuk representasi sempurna.
